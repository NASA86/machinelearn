```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
datasets=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
datasets
```

```python
indep=datasets[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=datasets[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['squared_error','absolute_error','friedman_mse','poisson'],'max_features': [None,'sqrt','log2'],'n_estimators':[10,100]}
from sklearn.ensemble import RandomForestRegressor
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
re=grid.cv_results_
print("The R_score value for best parameters:",format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
future_pred=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future prediction:",format(future_pred))
```
